```python
from fastai.vision.all import *
from ipywidgets import *
```

```python
learner = load_learner('./model.pkl')
```

```python
def refresh(button):
    displayer.value = uploader.data[0] if uploader.data[0] else ''
    
def classify(button):
    pred = learner.predict(uploader.data[0]) if uploader.data[0] else 'no img'
    label.value = pred[0]
    print(pred)
    
```

```python
uploader = widgets.FileUpload()
displayer = widgets.Image(width='300',heigth='300')
refresh_button = widgets.Button(description='Refresh')
refresh_button.on_click(refresh)
classify_button = widgets.Button(description='Classify')
classify_button.on_click(classify)
label = widgets.Label(value='Waiting')
```

```python
button_box = HBox([refresh_button, classify_button])
box = VBox([uploader, displayer,button_box,label])
```
